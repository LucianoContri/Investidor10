```python
import pandas as pd
        
investidor10_acoes = pd.read_csv('investidor10_acoes_1.csv', sep = ',')
```

```python
investidor10_acoes.info()
```

```python
# renomeando unnamed para indicadores
investidor10_acoes.rename(columns={'Unnamed: 0': 'indicadores'}, inplace=True)
```

```python
investidor10_acoes.nunique()
```

```python
investidor10_acoes.head()
```

```python
# tratando '-' como NaN
investidor10_acoes = investidor10_acoes.replace('-', pd.NA)
```

```python
investidor10_acoes.head()
```

```python
# tratando NaN como 0
investidor10_acoes = investidor10_acoes.fillna('0')
```

```python
investidor10_acoes.head()
```

```python
investidor10_acoes.info()
```

```python
for i in investidor10_acoes.columns[1:-1]:  # assuming 'acao' is the last column
    # Remove percentage signs
    investidor10_acoes[i] = investidor10_acoes[i].str.replace('%', '', regex=False)
    investidor10_acoes[i] = investidor10_acoes[i].apply(lambda x: '0' if x == '-' else x)
    investidor10_acoes[i] = investidor10_acoes[i].str.replace('.', '', regex=False)
        # Then, replace commas that are used as decimal separators with a period
    investidor10_acoes[i] = investidor10_acoes[i].str.replace(',', '.', regex=False)
investidor10_acoes.head(70)
```

```python
#transformando em float as colunas que são numéricas
for i in investidor10_acoes.columns[1:-1]:
    # achando caso tenha algum valor que não seja float
    for j in investidor10_acoes[i]:
        try:
            float(j)
        except:
            #achando o index do valor que não é float
            print( i, j, investidor10_acoes[i][investidor10_acoes[i] == j].index[0])
            break
    investidor10_acoes[i] = investidor10_acoes[i].astype(float)
```

```python
investidor10_acoes.head()
```

```python
investidor10_acoes.info()
```

```python
investidor10_acoes['mov_avg_2019_2022'] = investidor10_acoes[['2019','2020', '2021','2022']].mean(axis=1)
investidor10_acoes['mov_avg_2019_2023'] = investidor10_acoes[['2019','2020','2021','2022','2023']].mean(axis=1)
investidor10_acoes['mov_avg_2020_2023'] = investidor10_acoes[['2020','2021','2022','2023']].mean(axis=1)
investidor10_acoes['mov_avg_2021_2024'] = investidor10_acoes[['2021','2022','2023','Atual']].mean(axis=1)
# Now each indicator has a moving average column which can be used as a feature
```

```python
investidor10_acoes.head(100)
```

```python
pivot_df = investidor10_acoes.pivot_table(index='acao', columns='indicadores', values='mov_avg_2019_2022', aggfunc='first')
pivot_df.head()
```

```python
# rename index para acao
pivot_df = pivot_df.rename_axis("acao", axis=0)
```

```python
# selecionando linhas com acao abcb4
investidor10_acoes[investidor10_acoes['acao'] == 'abcb4']
```

```python
#selecionando a ação petr4 para pegar a linha
pivot_df.loc['abcb4']
```

```python
# tratando nulos como 0
pivot_df = pivot_df.fillna(0)
```

```python
pivot_df.info()
```

```python
# Crie uma série que mapeia cada ação ao seu valor 'ROIC' correspondente do ano 'Atual'
roic_atual = investidor10_acoes[(investidor10_acoes['indicadores'] == 'ROIC')].set_index('acao')['Atual']

# Adicione a série como uma nova coluna ao DataFrame pivotado
pivot_df['ROIC_Atual'] = pivot_df.index.map(roic_atual.to_dict())
```

```python
# Crie uma série que mapeia cada ação ao seu valor 'ROIC' correspondente do ano 'Atual'
roic_atual = investidor10_acoes[(investidor10_acoes['indicadores'] == 'ROIC')].set_index('acao')['2023']

# Adicione a série como uma nova coluna ao DataFrame pivotado
pivot_df['ROIC_2023'] = pivot_df.index.map(roic_atual.to_dict())
```
